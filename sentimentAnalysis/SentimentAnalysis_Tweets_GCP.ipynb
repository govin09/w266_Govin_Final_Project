```python
from google.cloud import language_v1
from google.cloud.language_v1 import enums
import pandas as pd
```

```python
df = pd.read_csv('sentimentTesting.csv')
df
```

```python
def sample_analyze_sentiment(text_content):
    """
    Analyzing Sentiment in a String

    Args:
      text_content The text content to analyze
    """

    client = language_v1.LanguageServiceClient()
    type_ = enums.Document.Type.PLAIN_TEXT
    language = "en"
    sentence = {"content": text_content, "type": type_, "language": language}
    encoding_type = enums.EncodingType.UTF8

    response = client.analyze_sentiment(sentence, encoding_type=encoding_type)
    #print(type(response))
    return response
```

```python
# To get a dataframe with the Sentiment Score and Magnitude
#Google recommends to not preprocess the text before using the API
listofseries = []
for index,row in df.iterrows():
    response = sample_analyze_sentiment(row['text'])
    row['Sentiment Score'] = round(response.document_sentiment.score,3)
    row['Sentiment Magnitude'] = round(response.document_sentiment.magnitude,3)
    listofseries.append(row)
dffinal = pd.DataFrame(listofseries)
dffinal
```

```python
dffinal.to_csv('Results-SA.csv')
```
